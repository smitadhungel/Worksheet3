```python
#total fare all paasengers are
total_fare =df['Fare'].sum()
print(f"Total Fare: {total_fare}")
```

```python
#total fare for each class
first_class_fare = df[df['Pclass'] == 1]['Fare'].sum()
second_class_fare = df[df['Pclass'] == 2]['Fare'].sum()
third_class_fare = df[df['Pclass'] == 3]['Fare'].sum()

print(f"First Class Fare: {first_class_fare}")
print(f"Second Class Fare: {second_class_fare}")
print(f"Third Class Fare: {third_class_fare}")
```

```python
#the proportion of total fare revenue for each class
first_class_percentage = (first_class_fare / total_fare) * 100
second_class_percentage = (second_class_fare / total_fare) * 100
third_class_percentage = (third_class_fare / total_fare) * 100

print(f"First Class Fare Percentage: {first_class_percentage:.2f}%")
print(f"Second Class Fare Percentage: {second_class_percentage:.2f}%")
print(f"Third Class Fare Percentage: {third_class_percentage:.2f}%")
```

```python
import numpy as np
conditions = [
    (df['Age'] < 18),
    (df['Age'] >= 18) & (df['Age'] < 65),
    (df['Age'] >= 65)
]
choices = ['child', 'adult', 'senior']

df['age_group'] =np.select(conditions, choices, default='adult')

df.head()

```

```python
# Count passngers in age group
age_group_counts = df['age_group'].value_counts()

# Total  passengers
total_passengers = len(df)

# Percent for each age group
age_group_percentage = (age_group_counts / total_passengers) * 100

# Dsiplay percentage for age groups
age_group_percentage
```
